# Web Scraping, a.k.a. `lyrics`

### ðŸŽ¯ Goal: scarpe lyrics from EMINEM and MADONNA

```python
import requests
import re
import time
from bs4 import BeautifulSoup

import pandas as pd
```

Requests allows us to send HTTP requests to website/servers. It sends back a response code, and the full html (if successful).

```python
# example with spiced academy

# avoid getting blocked 
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

url_mad = 'https://www.lyrics.com/artist/Madonna/64565'
```

```python
response_mad = requests.get(url_mad, headers=headers)
response_mad.text
```

```python
response_mad.status_code
```

# Starting with **Beautiful Soup**

```python
soup_mad = BeautifulSoup(response_mad.text, 'html.parser')
soup_mad
```

```python
soup_mad.head.title.text
```

## Getting all lyric titles in the html page

* FIRST: get individual reference first to see where it is

```python
soup_mad.find(class_="tal qx").a['href']
```

```python
# another way
#soup_mad.find('td', attrs={'class':'tal qx'}).text
```

```python
# use find_all for MADONNA
all_mado_titles = soup_mad.find_all(class_='tal qx')

#empty list to append all title names

titles_mad = []

for title in all_mado_titles:
    titles_mad.append(title.a.text)
```

```python
titles_mad
```

## Extract the song **links** 

```python
# individually 

"www.lyrics.com" + soup_mad.find('td', attrs={'class' : 'tal qx'}).a['href']
```

```python
#madonna 
links_mad = []

poemlinks = soup_mad.find_all('td', attrs={'class' : 'tal qx'})
for link in poemlinks:
    links_mad.append("http://www.lyrics.com" + link.a['href'])
links_mad
```

## Create DataFrame

```python
madonna = pd.DataFrame({'title': titles_mad, 'links':links_mad})
madonna
```

# Clean **MADONNA** df

```python
# shape --> 2913 rows Ã— 2 columns
madonna
```

```python
for i in range(len(madonna)):
    madonna.title[i] = madonna.title[i].split('[')[0]
    madonna.title[i] = madonna.title[i].split('(')[0]
    madonna.title[i] = madonna.title[i].lower()
```

```python
madonna['title'].value_counts()
```

```python
madonna.drop_duplicates(subset='title', inplace=True)
```

```python
madonna.reset_index(inplace=True, drop=True)
```

## Creating new column with lyrics

* df is cleaned from repeated songs
    * all lower case
    * no []
    * no ()

```python
# creating a new column first filled in with black
madonna[['song_lyric', 'artist']] = ['blank', '1']
```

```python
madonna    # shape --> 430 rows Ã— 3 columns
```

## Madonna

```python
for i in range(len(madonna.links)):
    try:
        lyric = BeautifulSoup(requests.get(madonna.links[i]).text)
        madonna['song_lyric'][i] = lyric.find('pre', attrs={'id':'lyric-body-text'}).text
        time.sleep(4)
    except:
        print('Something wrong with file: ', madonna.links[i])
        #print(exception, df.file[i])
        madonna['song_lyric'] = 'NaN'
        # question, there is a fuckton of generated NaN without exception generated...
        continue
```
